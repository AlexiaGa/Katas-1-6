## Ejercicio 1: Uso de ciclos while en Python

```python
new_planet = input ("Por favor, introduce el nombre de un planeta del sistema solar")
planets = [new_planet]

while new_planet.lower() != 'termine':
    new_planet = input("introduce un nuevo planeta, si has terminado por favor escribe la palabra 'Terminado'")
    if new_planet:
        planets.append(new_planet)
print (f"""La lista que acabas de crear es: {planets}""")
```

```python
planet = ""
for planet in planets:
    print (planet)
```
